```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
from tensorflow import keras
```

```python
df=pd.read_csv('Iris_dataset.csv')
```

```python
df.head()
```

```python
df['Species'].value_counts()
```

```python
df.isnull().sum()
```

```python
Y=pd.get_dummies(df.Species)
```

```python
Y
```

```python
X=df.drop(columns=['Id','Species'])
X.head()
```

```python
Y
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.25,shuffle=True,random_state=69)
```

```python
x_train
```

```python
y_train
```

```python
from keras import Sequential
from keras.layers import Dense,Dropout
```

```python
model=Sequential()
model.add(Dense(units=32,activation="relu"))
model.add(Dense(units=32,activation="relu"))
model.add(Dense(units=3,activation="softmax"))
```

```python
model.compile(optimizer='Adam',loss='categorical_crossentropy',metrics=['accuracy'])
model.summary()
```

```python
model.fit(x_train,y_train, epochs=100 ,batch_size=5)
```

```python
prediction=model.predict(x_test)
print(prediction)
```

```python
ans=model.evaluate(x_test,y_test)
ans
```

## Hence we have made a model with 97% accuracy which is able to classify the species of flower as [1,0,0] for iris_setosa [0,1,0] for iris_versicolor and [0,0,1] for iris_virginica
